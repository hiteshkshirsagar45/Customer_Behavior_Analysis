```python
import pandas as pd

customer_data = pd.read_csv('C:/Users/HP/Desktop/Customer_Analysis/customer_data.csv')

customer_data.head()
```

```python
customer_data.isnull().sum()
```

```python
customer_data['Review Rating'] = customer_data.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
customer_data.isnull().sum()
```

```python
customer_data.columns = customer_data.columns.str.lower()
customer_data.columns = customer_data.columns.str.replace(' ','_')
customer_data = customer_data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
customer_data.columns
```

```python
#create new column age_group
labels = ['Young_Adult','Adult','Middle_aged','Senior']
customer_data['age_group'] = pd.qcut(customer_data['age'], q=4, labels= labels)
```

```python
customer_data[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly': 14,
    'Annually' : 365,
    'Every 3 Months': 90
}

customer_data['purchase_frequency_days'] = customer_data['frequency_of_purchases'].map(frequency_mapping)
```

```python
customer_data[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
customer_data[['discount_applied','promo_code_used']].head(10)
```
